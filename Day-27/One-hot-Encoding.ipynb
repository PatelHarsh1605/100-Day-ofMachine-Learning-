# One-Hot Encoding

```python
# Whenever you have n colns then e generally create n-1 colns for classification


import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("cars.csv")
df.head()
```

```python
df['owner'].value_counts()
```

# One-hot Encoding With Pandas

```python
# One Hot Encoding Using Pandas

pd.get_dummies(df,columns=['fuel','owner'])

```

# One-Hot Encoding K-1

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

# OneHotEncoding With Sklearn

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
```

```python
X_test_new = ohe.transform(X_test[['fuel','owner']])
```
