```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13) # type: ignore
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Gradient Decent

y_pred = ((78.35*X)+0).reshape(4)
```

```python
m = 78.35
b = 0 

loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Finding the value of new intercept b using learning rate

lr = 0.1


step_size  = lr*loss_slope
step_size
```

```python
b = b - step_size
print(b)
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='green',label='b={}'.format(b))
plt.plot(X,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```
