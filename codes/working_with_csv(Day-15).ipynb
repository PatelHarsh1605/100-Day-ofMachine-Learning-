```python
# Convertors

def rename(name) :
    if name == "Royal Challengers Bangalore" :
        return "RCB"
    elif name == "Kings XI Punjab" :
        return "PBKS"
    elif name == "Chennai Super Kings" :
        return "CSk"
    elif name == "Rajasthan Royals" :
        return "RR"
    elif name == "Mumbai Indians" :
        return "MI"
    elif name == "Delhi Capitals" :
        return "DC"
    elif name == "Sunrisers Hyderabad" :
        return "SRH"
    elif name == "Kolkata Knight Riders" :
        return "KKR"
    else :
        return name
    
df = pd.read_csv(r"C:\Users\Admin\Downloads\IPL Matches 2008-2020.csv",converters={'team1' : rename,'team2' : rename})
df
```
